```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data=pd.read_csv('tripadvisor_hotel_reviews.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase']=data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords=stopwords.words('english')
```

```python
en_stopwords.remove("not")
```

```python
data["review_no_stopwords"]=data['review_lowercase'].apply(lambda x:" ".join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_puct']=data.apply(lambda x: re.sub(r"[*]","star",x['review_no_stopwords']),axis=1)
```

```python
data['review_no_stopwords_no_puct'][2]
```

```python
data.head()
```

```python
data['review_no_stopwords_no_puct']=data.apply(lambda x: re.sub(r"([^\w\s])","",x['review_no_stopwords_no_puct']),axis=1)
```

```python
data['tokenized']=data.apply(lambda x:word_tokenize(x['review_no_stopwords_no_puct']),axis=1)
```

```python
data['tokenized'][0]
```

```python
ps=PorterStemmer()
data['stemmed']=data['tokenized'].apply(lambda tokens :[ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
data['lemmatized']=data['tokenized'].apply(lambda tokens :[lemmatizer.lemmatize(token) for token in tokens])
```

```python
data.head()
```

```python
token_clean=sum(data['lemmatized'],[])
```

```python
unigrams= (pd.Series(nltk.ngrams(token_clean,1)).value_counts())
print(unigrams)
```

```python
biggrams= (pd.Series(nltk.ngrams(token_clean,2)).value_counts())
print(biggrams)
```
