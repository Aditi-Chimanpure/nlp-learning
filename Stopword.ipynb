```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords=stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
sentence="it was too far to go to the shop and he did not want her to walk"
```

```python
sentence_with_no_stop_words=' '.join(word for word in sentence.split() if word not in en_stopwords )
print(sentence_with_no_stop_words)
```

```python
en_stopwords.remove('was')
en_stopwords.remove('to')
en_stopwords.append('go')
```

```python
sentence="it was too far to go to the shop and he did not want her to walk"
sentence_with_no_stop_words_custom=' '.join(word for word in sentence.split() if word not in en_stopwords )
print(sentence_with_no_stop_words_custom)
```
