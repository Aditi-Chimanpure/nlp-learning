```python
# Dictionary of food items with their prices
prices = {
    "box_of_spaghetti" : 4,
    "lasagna"  : 5,
    "hamburger" : 2
   }
# Dictionary of food items with their quantities
quantity = {
    "box_of_spaghetti" : 6,
    "lasagna"  : 10,
    "hamburger" : 0
    }

# Initialize the total money spent to zero
money_spent = 0

# Loop through each item in the prices dictionary
for i in prices:
    # Calculate the cost for each item (price × quantity) and add it to the total
    money_spent = money_spent + (prices[i]*quantity[i])
    
# Display the total money spent
print (money_spent)
```
